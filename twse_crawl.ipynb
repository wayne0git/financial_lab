```
import numpy as np
import pandas as pd
import requests

from io import StringIO
```

```
# Parameter
DATESTR = '20180131'
```

```
# Download
r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + DATESTR + '&type=ALL')
```

```
# Text to DataFrame
text = r.text.replace("=", "")  # Remove = in "證券代號"
header_index = ["證券代號" in l for l in r.text.split("\n")].index(True) - 1
df = pd.read_csv(StringIO(text), header=header_index)
```

```
df.apply(lambda s: pd.to_numeric(s.astype(str).str.replace(",", "").replace("+", "1").replace("-", "-1"), errors='coerce'))
df
```

```
import json
import numpy as np
import pandas as pd
import requests
```

### Parameter

```
DATE = '20180131'
STOCK_NO = '2330'
URL = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=%s&stockNo=%s' % (DATE, STOCK_NO)
```

### Download stock price (From https://www.twse.com.tw)

```
data = requests.get(URL).text
```

### Preprocessing

```
json_data = json.loads(data)
Stock_data = json_data['data']
StockPrice = pd.DataFrame(Stock_data, columns = ['Date','Volume','Volume_Cash','Open','High','Low','Close','Change','Order'])
```
