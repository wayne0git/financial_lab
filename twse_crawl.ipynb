```
import numpy as np
import pandas as pd
import requests

from io import StringIO
```

```
# Parameter
DATESTR = '20180131'
```

### Download stock price (From https://www.twse.com.tw)

```
# Download
r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + DATESTR + '&type=ALL')
```

### Preprocessing

```
# Text to DataFrame
text = r.text.replace("=", "")  # Remove = in "證券代號"
header_index = ["證券代號" in l for l in r.text.split("\n")].index(True) - 1
df = pd.read_csv(StringIO(text), header=header_index)
```

```
df.apply(lambda s: pd.to_numeric(s.astype(str).str.replace(",", "").replace("+", "1").replace("-", "-1"), errors='coerce'))
df
```
